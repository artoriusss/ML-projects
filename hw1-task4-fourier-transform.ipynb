## Task 1. Continuous signals, pure tones, and Fourier series (1 pts)

An analog audiosignal $\mathbf{x}$ of duration $T$ is a real-valued (usually continuous) function over $[0,T]$, whose value $\mathbf{x}(t)$ at a time moment $t$ represents the amplitude of air pressure variations, i.e., the change in air pressure with respect to some reference value. These pressure variations cause virbations of the ear drums, which are converted into electromagnetic impulses that are then sent to our brain.

Physical laws of gas dynamics imply that each such signal can be represented as a superposition of periodic ***pure tones*** $\sin( \omega t + \phi)$, i.e.,
$$\mathbf{x}(t) = \sum a_k \sin(\omega_k t + \phi_k).$$
Here $a_k$ is the amplitude of the respective pure wave of ***frequency*** $\omega_k/2\pi$ Hz (Hertz, cycles per second) and phase shift of $\phi_k$ radians. Observe that a wave of frequency $\omega_k/2\pi$ Hz has a period of $2\pi/\omega_k$ seconds.

The signal $\mathbf{x}$ can be regarded as periodic of period $T$ and can be represented via the ***Fourier series***
$$
  \mathbf{x}(t) = \frac{a_0}{2} + \sum_k a_k \cos(2\pi k t / T) + \sum_k b_k \sin(2\pi k t/T),
$$
where
$$
 a_k:= \frac{2}{T} \int_0^T f(s)\cos(2\pi ks/T)\,ds,  \qquad b_k:= \frac{2}{T} \int_0^T f(s)\sin(2\pi ks/T)\,ds
$$
are the ***Fourier coefficients***.

```python
def signal(s):
  # ========= YOUR CODE STARTS HERE ========= #
  return 0.5 * np.sin(2*s) - np.sin(3*s)
  # ========== YOUR CODE ENDS HERE ========== #
N = 200

t = np.linspace(0, 4*np.pi, N+1)[:-1]
x = signal(t)
sigma = np.sqrt(0.01)

# ========= YOUR CODE STARTS HERE ========= #
noise = np.random.normal(loc=0, scale=sigma, size=N)
x_noise = x + noise
# ========== YOUR CODE ENDS HERE ========== #
```

```python
plt.plot(t, x_noise, label = "Noised signal")
plt.plot(t, x, "r-.", label = "Smooth signal")
plt.title('Original signal $\mathbf{x}$ and $\mathbf{x}_\mathrm{noise}$')
plt.xlabel("time $t$ (seconds)")
plt.ylabel("amplitude")
plt.legend()
plt.show()
```

```python
########################################
#     Trigonometric Fourier coefficients
########################################

rng = np.arange(N).reshape(1, -1)

# cos_samples and sin_samples are N x N matrices with sampled cosines/sines written in rows

# ========= YOUR CODE STARTS HERE ========= #

t_col = t.reshape(-1, 1) 
cos_samples = np.cos((2 * np.pi * rng * t_col) / T)
sin_samples = np.sin((2 * np.pi * rng * t_col) / T)

# ========== YOUR CODE ENDS HERE ========== #
```

```python
########################################
#  noiseless case
########################################

# ========= YOUR CODE STARTS HERE ========= #

a = np.mean(x.reshape(1, -1) * cos_samples, axis=1)    #  the sequence of N cosine coefficients of signal x
b = np.mean(x.reshape(1, -1) * sin_samples, axis=1)    #  the sequence of N  sine  coefficients of signal x

# ========== YOUR CODE ENDS HERE ========== #

plt.plot(range(N), b.T, "r", lw = 0.75, label = "Sine coeff")
plt.plot(range(N), a.T, "b", lw = 0.75, label = "Cosine coeff")
plt.xlabel("index k")
plt.ylabel("Coefficients")
plt.title('Trigonometric Fourier coefficients $a_k$ and $b_k$ for $\mathbf{x}$')
plt.legend()
plt.show()
```

Our signal does not contain any cosine terms, so all the cosine coefficients are about $0$ and blue line is just a horizontal line. With sine, coefficients, however, we can see a few jumps in our signal. Our signal comprises two sine waves, $\sin(2t)$ and $\sin(3t)$. The two jumps we see at around $k=3$ and $k=5$ correspond to the frequencies of sine waves. The jump visible at $k\approx 180$ can be due to aliasing.

```python
########################################
#  noisy signal
########################################

# ========= YOUR CODE STARTS HERE ========= #

a_noise =  np.mean(x_noise.reshape(1, -1) * cos_samples, axis=1)   #  the sequence of N cosine coefficients of noisy signal x_noise
b_noise =  np.mean(x_noise.reshape(1, -1) * sin_samples, axis=1)   #  the sequence of N  sine  coefficients of noisy signal x_noise

# ========== YOUR CODE ENDS HERE ========== #

plt.plot(range(N), b_noise.T, "r", lw = 0.75, label = "Sine coeff")
plt.plot(range(N), a_noise.T, "b", lw = 0.75, label = "Cosine coeff")
plt.xlabel("index k")
plt.ylabel("Coefficients")
plt.title('Trigonometric Fourier coefficients $a_k$ and $b_k$; noisy signal')
plt.legend()
plt.show()
```

Here we see that adding noise to the signal introduces small random variations in both Fourier coefficients at all frequencies, but the major features of the Fourier transform (the peaks at the main frequencies of the signal) actually remain about the same. This is because the noise should not affect the dominant frequencies of the signal and that is what we observe on the plot

```python
# Find the indices of the 6 largest |a_k| and |b_k| 
max_a_indices = np.argsort(np.abs(a))[::-1][:6]
max_b_indices = np.argsort(np.abs(b))[::-1][:6]

sampling_rate = N / t[-1]

frequencies_a = max_a_indices * sampling_rate / N
frequencies_b = max_b_indices * sampling_rate / N

magnitudes_a = np.abs(a[max_a_indices])
magnitudes_b = np.abs(b[max_b_indices])

print("6 most important a_k coefficients and their frequencies: ")
for idx, freq, mag in zip(max_a_indices, frequencies_a, magnitudes_a):
    print(f"k={idx}, |a_k|={mag:.3f}, f={freq:.3f}")

print("\n6 most important b_k coefficients and their frequencies: ")
for idx, freq, mag in zip(max_b_indices, frequencies_b, magnitudes_b):
    print(f"k={idx}, |b_k|={mag:.3f}, f={freq:.3f}")
```

```python
# Reconstruction of the signal using n sine and cosine Fourier coefficients
# Pay special attention to a_0!
def x_rec(n, a, b):
  # n is the number of coefficients a_k and b_k to be included
  # a is the 1d array of cosine coefficients
  # b is the 1d array of  sine  coefficients

  # ========= YOUR CODE STARTS HERE ========= #
    t = np.linspace(0, 4*np.pi, N)

    # Initialize the reconstructed signal
    x_hat = np.zeros_like(t)

    # Add the cosine terms to the reconstructed signal
    for k in range(n):
        x_hat += a[k] * np.cos(2 * np.pi * k * t / T)

    # Note: For the sine terms, start from k=1 because sin(0) = 0
    for k in range(1, n):
        x_hat += b[k] * np.sin(2 * np.pi * k * t / T)

    return x_hat
  # ========== YOUR CODE ENDS HERE ========== #
```

```python
def max_abs_difference(x_noise, x_rec):
    return np.max(np.abs(x_noise - x_rec))

n_values = [5, 10, 100, 200]
plt.figure(figsize=[10, 8])

for idx, n in enumerate(n_values):
    x_approx = x_rec(n, a_noise, b_noise)
    
    max_diff = max_abs_difference(x_noise, x_approx)
    
    plt.subplot(2, 2, idx+1)
    plt.plot(t, x_noise, label="Original", lw=2)
    plt.plot(t, x_approx.T, label=f"Reconstructed n={n}", linestyle="--")
    plt.title(f"n={n}, Max difference = {max_diff:.3f}")
    plt.xlabel("Time")
    plt.ylabel("Signal")
    plt.grid()
    plt.legend()

plt.tight_layout()
plt.show()
```

```python
np.allclose(x_noise, x_rec(200, a_noise, b_noise), rtol=0.01)
```

```python
np.max(np.abs(x_noise - x_rec(200, a_noise, b_noise)))
```

Also, important fact to consider is tha the improvement of reconstruction depends on how many of *the most important coefficients* we include into our function `x_rec`. From the previous task, we remember that among $b_k$ coefficients, the magnitude of $k=6$ was $0.5$. That is, it contributes very significantly to overall signal. This also means that including this term in our Fourier series approximation is very important to capture the primary characteristics of the signal. And that's why the second plot with $n=10$ (that includes 6th$b_k$) is already demonstrating that our approximation is significantly better.

However, for $n=200$, the reconstruction gets poor due to the overfitting. Including too many high-frequency Fourier coefficients causes the approximation to fit the noise in the signal, rather than the underlying trend.

This illustrates a key aspect of Fourier series approximation - the importance of choosing the right number of terms to include. Including the most significant coefficients in the series can yield a reasonable approximation of the signal, even with a limited number of terms, but the balance should be kept in order to avoid overfitting.

## Task 2. Sampling and aliasing (0.7 pts)

There are several reasons for appearance of the spurious frequencies in the above example, the main one being related to ***aliasing***.

Assume we measure an analogue signal $\mathbf{x}$ of duration $T$ on the time lattice $t_j = \Delta t \cdot j$, $j=0,1, \dots, N-1$. The time step $\Delta t = T/N$ is called the ***sample period***, and the number of samples per 1 second, $f_s = 1/\Delta t = N/T$, is the ***sampling rate***. The process of measuring the input analog signal at the points $t_j$ is called ***sampling***; we then discretize the obtained values $\mathbf{x}(t_j)$ to form a ***digital signal*** $\mathbf{x} = \bigl(x(0), x(t_1),\dots, x(t_{N-1})\bigr)$ of length $N$.

The point is, some pure tones (harmonics) $\sin(\omega t)$ or $\cos(\omega t)$ have identical samples. Indeed, to have
$$
  \cos(\omega_1 t_j) = \cos(\omega_2 t_j), \qquad j = 0,1, \dots, N-1,
$$
it is necessary and sufficient that
$$
  \sin\Bigl(\frac{\omega_1 - \omega_2}{2}t_j\Bigr) \sin\Bigl(\frac{\omega_1 + \omega_2}{2}t_j\Bigr) = 0
$$
i.e.,
$$
  (\omega_1 + \omega_2) \Delta t = 2\pi k \quad\text{or}\quad
  (\omega_1 - \omega_2) \Delta t = 2\pi k
$$
A similar condition can be written for two sinusoids $\sin(\omega_1 t)$ and $\sin(\omega_2 t)$ to have identical samples.

If we take 20 samples of a signal with a duration of $4\pi$, the time step $\Delta t$ is $T/N = 4\pi/20 = 0.2\pi$. Therefore, the sampling rate $f_s$ (the number of samples per 1 second) is $1/\Delta t = 1/(0.2\pi) = 5/\pi = 0.2 \pi$ samples per second.


We need to find $k$ so that the samples of $-\sin(kt)$ and $\sin(2t)$ are the same. The condition for two tones to have identical samples is:

$$(\omega_1 - \omega_2) \Delta t = 2\pi n$$ or $$(\omega_1 + \omega_2) \Delta t = 2\pi n$$

for some integer $n$.

Given that $\omega_1 = 2$ and $\omega_2 = -k$, we can have $(2 + k) \Delta t = 2\pi n$

Solving for $k$ with $\Delta t = \dfrac{T}{N} = \dfrac{4\pi}{20} = 0.2\pi$ and considering the periodicity of the sine function, we choose $n=2$ to satisfy the equation, given that the period of $\sin$ is $2\pi$.

So, substituting $\Delta t = 0.2\pi$ and $n=2$ into $(2 + k) \Delta t = 2\pi n$, we get:

$$2 + k = 10 \Longrightarrow k = 18$$

Due to the property of the sine function being periodic, we can subtract multiples of the sampling rate ($f_s = 5$) from $k$ to get another $k$ that will also satisfy the condition. So, we can subtract $f_s = 5$ twice from $k = 18$, which leads us to $k = 8$ (subtracting three times would not work because the resulting $k=3$ would not satisfy the equation).

So, the smallest positive $k$ that will make $-\sin(kt)$ coincide with $\sin(2t)$ at the 20 sample points is $k = 8$.

## Task 3. Fourier transform and FFT (1 pt)

Another reason why spurious frequencies appeared in ***Task 1.2*** is due to integration error. When we replaced the integrals for calculating the Fourier coefficinets $a_k$ and $b_k$ by the Riemann integral sums, aliasing for sinusoids with high frequencies led to results identical with those for smaller frequencies, although the actual $a_k$ and $b_k$ are equal to zero for smooth signal and are very small for noisy signal.

In other words, when replacing integrals with sums, we actually moved the object of study from the analog signals in $L_2[0,4\pi]$ to digital signals in $\mathbb{R}^N$. Certainly, $N$ sampled cosines and $N$ sampled sines formed linearly dependent vectors in $\mathbb{R}^N$, and that is why using all of them to reconstruct the original signal $\mathbf{x}_\mathrm{noise}$ failed.

When working with digital signals in $\mathbb{R}^N$, we only need $N$ basis vectors. Observe that sampled cosines are <font color="red">*even*</font> functions with respect to mid-point, i.e., $\cos(t_j) = \cos(t_{N-j})$; likewise, sines are <em><font color="red">odd</font></em> with respect to the mid-point. Another observation is that neither sampled cosines nor sampled sines form orthogonal sets. For that reason, a more convenient orthogonal basis in $\mathbb{C}^N$ is given by the sampled exponentials $\mathrm{e}^{2\pi \mathrm{i} k t/T}$, $k=0,1,\dots,N-1$: $$\mathbf{e}_k:=\bigl(1, \mathrm{e}^{2\pi \mathrm{i}k t_1/T}, \dots, \mathrm{e}^{2\pi \mathrm{i}t_{N-1}/T}\bigr)$$
Observe that the exponential function $\mathrm{e}^{2\pi \mathrm{i} k t/T}$ is a pure complex harmonics of frequency $k/T$ Hz.

Every vector $\mathbf{x} \in \mathbb{C}^N$ can be represented in this basis in the standard way: $$\mathbf{x} = \sum_k c_k \mathbf{e}_k, \qquad c_k = \frac{\langle \mathbf{x}, \mathbf{e}_k\rangle}{\langle \mathbf{e}_k, \mathbf{e}_k\rangle} = \frac1N\sum_j x_j \mathrm{e}^{- 2\pi \mathrm{i}t_{j}/T} \tag{4}$$
The $c_k$ are <font color = "red">complex Fourier coefficients</font> of $\mathbf{x}$, and the mapping
$$
   \mathbf{x} \mapsto (c_k)_{k=0}^{N-1}
$$
is the <font color = "red">Fourier transform</font>. While the direct approach of calculating $c_k$ requires $O(N^2)$ flops, the ***Fast Fourier transform (FFT)*** performs this in $O(N\log N)$ flops. Given that a standard audio signal of duration 1 second is sampled $N = 44\,100$ times, FFT gives a significant gain in time efficiency.
